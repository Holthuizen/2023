```python
#times = np.array([7,15,30]) #time available
#distances = np.array([9,40,200]) #record distances
times = np.array([58,99,64,69]) #time available
distances = np.array([478,2232, 1019, 1071]) #record distances
"""Your toy boat has a starting speed of zero millimeters per millisecond. 
For each whole millisecond you spend at the beginning of the race holding down the button, the boat's speed increases by one millimeter per millisecond."""

#for each pair find the number of ways to beat the record.
answer = 1
for i in range(len(times)): #for all races
    winning_strats = 0
    for b in range(times[i]):
        dist =  b * (times[i] - b)
        print(times[i], distances[i], b, dist)
        if dist > distances[i]:
            winning_strats += 1
    answer *= winning_strats

answer
```
