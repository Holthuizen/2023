```python
import sys, math 
import numpy as np
sys.setrecursionlimit(30000)# Set the maximum recursion depth to 30k
```

```python
"""Determine the ASCII code for the current character of the string.
Increase the current value by the ASCII code you just determined.
Set the current value to itself multiplied by 17.
Set the current value to the remainder of dividing itself by 256."""

def hash(message, current_value = 0): 
    for l in message:
        current_value += ord(l)
        current_value *=  17
    return current_value % 256
```

```python
boxes = np.empty(256, dtype=object)
for i in range(len(boxes)):
    boxes[i] = {}
```

```python
def parse_command(instruction):
    if instruction[-1].isdigit():
        return instruction.split("="), 1
    else: 
        return instruction.split("-"), -1
```

```python
for instruction in "".join(open("input.txt")).split(','):
    instruction, operation = parse_command(instruction)
    print(instruction[0], hash(instruction[0]), operation)
    if operation > 0:
        boxes[hash(instruction[0])][instruction[0]] = instruction[1]

        #boxes[hash(instruction[0])].remove()
```
